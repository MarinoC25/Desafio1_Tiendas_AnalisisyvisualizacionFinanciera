```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. An√°lisis de facturaci√≥n



Agregar una columna para identificar la tienda

```python
tienda["Tienda"] = "Tienda 1"
tienda2["Tienda"] = "Tienda 2"
tienda3["Tienda"] = "Tienda 3"
tienda4["Tienda"] = "Tienda 4"
```

```python
tienda.head()
```

Unir todos los DataFrames

```python
df = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)
```

Verificar estructura

```python
print(df.head())
print(df.info())
```

Revisi√≥n de datos

```python
print(df.describe())     # Estad√≠sticas de columnas num√©ricas
print(df.isnull().sum()) # Valores nulos por columna
```

Facturaci√≥n total consolidada

```python
facturacion_total = df["Precio"].sum()
print(f"üí∞ Facturaci√≥n total consolidada: ${facturacion_total:,.2f}")
```

Facturaci√≥n por tienda

```python
facturacion_tienda = df.groupby("Tienda")["Precio"].sum().reset_index()
print("\nüí∞ Facturaci√≥n por tienda:")
print(facturacion_tienda)
```

 Facturaci√≥n por categor√≠a y tienda

```python
facturacion_categoria_tienda = df.groupby(["Tienda", "Categor√≠a del Producto"])["Precio"].sum().reset_index()
print("\nüìä Facturaci√≥n por categor√≠a y tienda:")
print(facturacion_categoria_tienda)
```

Facturaci√≥n por m√©todo de pago y tienda

```python
facturacion_pago_tienda = df.groupby(["Tienda", "M√©todo de pago"])["Precio"].sum().reset_index()
print("\nüí≥ Facturaci√≥n por m√©todo de pago y tienda:")
print(facturacion_pago_tienda)
```

Ticket promedio por tienda

```python
ticket_promedio = df.groupby("Tienda")["Precio"].mean().reset_index()
ticket_promedio.columns = ["Tienda", "Ticket Promedio"]
print("\nüéü Ticket promedio por tienda:")
print(ticket_promedio)
```

# Facturaci√≥n por M√©todo de Pago en cada Tienda

```python
import plotly.express as px

# Agrupar datos por tienda y m√©todo de pago
facturacion_pago_tienda = df.groupby(["Tienda", "M√©todo de pago"])["Precio"].sum().reset_index()
facturacion_pago_tienda.rename(columns={"Precio": "Facturaci√≥n"}, inplace=True)

# Crear gr√°fico de barras agrupadas
fig = px.bar(
    facturacion_pago_tienda,
    x="Tienda",
    y="Facturaci√≥n",
    color="M√©todo de pago",
    barmode="group",
    text="Facturaci√≥n",
    title="üí≥ Facturaci√≥n por M√©todo de Pago y Tienda",
    labels={"Facturaci√≥n": "Facturaci√≥n ($)", "Tienda": "Tiendas", "M√©todo de pago": "M√©todo de Pago"}
)

# Formato de etiquetas de texto
fig.update_traces(
    texttemplate='$%{text:,.0f}',  # Etiquetas con formato monetario
    textposition="outside"
)

# Ajustar dise√±o
fig.update_layout(
    title_font=dict(size=22),
    yaxis_title="Facturaci√≥n ($)",
    xaxis_title="Tienda",
    template="plotly_white",
    bargap=0.3,
    yaxis=dict(showgrid=True, gridcolor="lightgray")
)

fig.show()
```

**Conclusi√≥n:**

Aunque las diferencias no son extremas, existe margen para reducir los costos de env√≠o en las tiendas con valores m√°s altos y as√≠ mejorar el margen de ganancia general.

```python
import pandas as pd

# Datos de ejemplo (puedes reemplazarlos por tus datos reales)
data = {
    "Tienda": ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"],
    "Facturacion": [1150880000, 1116344000, 1098020000, 1038376000],
    "Ticket_Promedio": [487867.91, 473227.43, 465459.77, 440362.89],
    "Envio_Promedio": [26018.61, 25216.24, 24805.68, 23459.46],
    "Calificacion": [3.98, 4.04, 4.05, 4.00]
}

df = pd.DataFrame(data)

# Ponderaciones de cada criterio
pesos = {
    "Facturacion": 0.4,
    "Ticket_Promedio": 0.15,
    "Envio_Promedio": 0.10,  # Menor es mejor, se invertir√° el valor
    "Calificacion": 0.15,
    "Rentabilidad": 0.20  # Si no tienes rentabilidad, puedes eliminarla o asumir un valor ficticio
}

# Normalizar datos (para que todas las m√©tricas est√©n en una misma escala)
for col in ["Facturacion", "Ticket_Promedio", "Envio_Promedio", "Calificacion"]:
    if col == "Envio_Promedio":
        # Para los costos de env√≠o, mientras m√°s bajo mejor ‚Üí invertimos el valor
        df[col + "_norm"] = 1 - (df[col] - df[col].min()) / (df[col].max() - df[col].min())
    else:
        df[col + "_norm"] = (df[col] - df[col].min()) / (df[col].max() - df[col].min())

# Calcular √≠ndice ponderado
df["Indice_Desempeno"] = (
    df["Facturacion_norm"] * pesos["Facturacion"] +
    df["Ticket_Promedio_norm"] * pesos["Ticket_Promedio"] +
    df["Envio_Promedio_norm"] * pesos["Envio_Promedio"] +
    df["Calificacion_norm"] * pesos["Calificacion"]
)

# Ordenar de mejor a peor
df = df.sort_values("Indice_Desempeno", ascending=False)

# Mostrar resultado
display(df)
```

# Indice de Desempe√±o

```python
import plotly.express as px

# Crear gr√°fico de barras
fig = px.bar(
    df,
    x="Tienda",
    y="Indice_Desempeno",
    text="Indice_Desempeno",
    title="√çndice de Desempe√±o por Tienda",
    labels={"Indice_Desempeno": "√çndice de Desempe√±o (%)"},
    color="Tienda"
)

# Mostrar el texto como porcentaje con 1 decimal
fig.update_traces(texttemplate='%{y:.1%}', textposition='outside')

# Ajustar rango del eje Y (0% a 100%)
fig.update_yaxes(range=[0, 1], tickformat=".0%")

fig.show()

```
